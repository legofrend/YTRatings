```python
import json

# Открываем файл для чтения
filename = 'logs/data.json'
# filename = 'frontend/src/assets/data.json'
with open(filename, 'r', encoding='utf-8') as file:
    # Загружаем данные из файла
    data = json.load(file)
    data = data[0]
print(data["period"], data["category_id"])

for item in data["data"]:
    for key in ["score", "score_change"]:
        item[key + "_display"] = round(item["stat"][key]/10**6)
    item["score_video_display"] = round(item["top_videos"][0]["stat"]["score"]/10**6)
    
with open("video_gen/script_tmpl.txt", 'r', encoding='utf-8') as f:
    tmpl = f.read()

script = tmpl.format(data = data["data"])    
with open(data["period"] + "_script.txt", 'w', encoding='utf-8') as f:
    f.write(script)
```
